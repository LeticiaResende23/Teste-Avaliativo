```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Definir a URL da API
url = 'https://contratos.comprasnet.gov.br/api/contrato/ug/113601'


```

```python
# Fazer a requisição GET e verificar o status
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    # Converter os dados para um DataFrame do pandas
    df = pd.DataFrame(data)
    display(df.head())  # Exibir as primeiras linhas do DataFrame
else:
    print(f'Erro na requisição: {response.status_code}')


```

```python
# Filtrar apenas colunas numéricas para análise
df_numeric = df.select_dtypes(include=['number'])

# Análise descritiva
descriptive_stats = df_numeric.describe()

# Exibir as estatísticas descritivas
print("Estatísticas Descritivas:")
display(descriptive_stats)
```

```python
# Calcular as médias de todos os principais indicadores
means = df_numeric.mean()

# Criar um gráfico de barras para as médias
plt.figure(figsize=(10, 6))
means.plot(kind='bar', title='Médias dos Indicadores Quantitativos', grid=True)
plt.xlabel('Indicadores')
plt.ylabel('Média')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Verificar valores nulos em cada coluna
missing_data = df.isnull().sum()
print("Valores Nulos por Coluna:")
print(missing_data)

# Identificar colunas que contêm objetos complexos
non_hashable_cols = df.applymap(lambda x: isinstance(x, dict)).any()
non_hashable_cols = non_hashable_cols[non_hashable_cols].index.tolist()
print(f"Colunas com objetos complexos: {non_hashable_cols}")

# Remover temporariamente as colunas com objetos complexos para verificar duplicatas
df_simple = df.drop(columns=non_hashable_cols, errors='ignore')

# Verificar duplicatas nas colunas simples
duplicates = df_simple.duplicated().sum()
print(f"Total de registros duplicados (colunas simples): {duplicates}")

# Verificar distribuição de categorias em colunas categóricas simples
categorical_columns = df_simple.select_dtypes(include=['object']).columns
categorical_summary = {}

for col in categorical_columns:
    categorical_summary[col] = df_simple[col].value_counts()


```
